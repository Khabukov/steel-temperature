- Металлургический комбинат «Стальная птица» реализует проект по оптимизации производственных расходов. Одним из ключевых направлений проекта является снижение потребления электроэнергии на этапе обработки стали.

- Для достижения этой цели необходимо разработать модель прогнозирования температуры сплава. Разработанная модель будет использоваться для имитационного моделирования технологического процесса, что позволит эффективно контролировать температурный режим и оптимизировать энергопотребление.

- Успешная реализация проекта позволит комбинату существенно сократить производственные расходы при сохранении высокого качества продукции.

- Критерии оценки: Значение метрики MAE <= 6,8.

**План реализации проекта**

1. Загрузка и первичный анализ данных

3. Объединение данных
- Определение ключа для объединения датафреймов
- Объединение отобранных признаков в единый датафрейм
- Проверка корректности объединения

4. Исследовательский анализ и предобработка данных объединённого датафрейма
- Визуализация распределений признаков
- Корреляционный анализ
- Генерация новых признаков на основе имеющихся
- Дополнительная очистка данных

5. Подготовка данных и обучение модели
- Разделение на обучающую и тестовую выборки
- Масштабирование числовых признаков
- Кодирование категориальных переменных
- Учет специфики данных для различных моделей
- Обучение минимум двух различных моделей
- Подбор гиперпараметров для одной из моделей (минимум два параметра)
- Оценка качества моделей на валидационной выборке
- Финализация модели

6. Выбор лучшей модели
- Оценка качества на тестовой выборке
- Анализ ошибок модели

### Описание данных

Данные состоят из нескольких датафреймов, полученных из разных источников:

- **data_arc** — данные об электродах;
- **data_bulk** — данные о подаче сыпучих материалов (объём);
- **data_bulk_time** — данные о подаче сыпучих материалов (время);
- **data_gas** — данные о продувке сплава газом;
- **data_temp** — результаты измерения температуры;
- **data_wire** — данные о проволочных материалах (объём);
- **data_wire_time** — данные о проволочных материалах (время).

Датафрейм **data_arc**

- **`key`** — номер партии;
- **`arc_heating_start`** — время начала нагрева;
- **`arc_heating_end`** — время окончания нагрева;
- **`active_power`** — значение активной мощности;
- **`reactive_power`** — значение реактивной мощности.

Датафрейм **data_bulk**

- **`key`** — номер партии;
- **`bulk_1 … bulk_15`** — объём подаваемого материала.

Датафрейм **data_bulk_time**

- **`key`** — номер партии;
- **`bulk_time_1 … bulk_time_15`** — время подачи материала.

Датафрейм **data_gas**

- **`key`** — номер партии;
- **`gas_1`** — объём подаваемого газа.

Датафрейм **data_temp**

- **`key`** — номер партии;
- **`measurement_time`** — время замера;
- **`temperature`** — значение температуры.

Датафрейм **data_wire**

- **`key`** — номер партии;
- **`wire_1 … wire_9`** — объём подаваемых проволочных материалов.

Датафрейм **data_wire_time**

- **`key`** — номер партии;
- **`wire_time_1 … wire_time_9`** — время подачи проволочных материалов.

Во всех датафреймах столбец **`key`** содержит номер партии. В датафреймах может быть несколько строк с одинаковым значением **`key`**: они соответствуют разным итерациям обработки.

Создадим функцию `calc_diff_time`, которая будет определять разницу во времени между соседними столбцами в датафрейме.

```python
def calc_diff_time(data):
    columns = data.columns.difference({'key'})
    
    new_columns = []
    
    for i in range(len(columns) - 1):
        current_column = columns[i]
        next_column = columns[i + 1]
        
        # Вычисляем разницу во времени
        time_diff = data[current_column] - data[next_column]
        
        # Преобразуем разницу в секунды и сохраняем в новый столбец
        new_column_name = f'{current_column}_diff'
        data[new_column_name] = time_diff.dt.total_seconds().abs()
        
        # Добавляем название нового столбца в список
        new_columns.append(new_column_name)
    
    # Удаляем старые столбцы
    data = data.drop(columns=columns)
    
    # Оставляем только новые созданные столбцы и 'key'
    data = data[['key'] + new_columns]
    
    return data
```

### RandomForestRegressor() и CatBoostRegressor()

Для подбора параметров `RandomForestRegressor()` и `CatBoostRegressor()` и предобработки данных создадим пайплайн.

```python
#создаём списки с названиями признаков

num_columns = [
    'duration',
    'bulk_sum',
    'gas_1',
    'start_temp',
    'wire_sum',
    'mean_bulk_time',
    'energy',
    'full_power',
    'mean_wire_time'
]
```

```python
# создаём пайплайн для числовых признаков с обработкой пропусков
num_pipe = Pipeline([
    ('simpleImputer_num', SimpleImputer(missing_values=np.nan, strategy='median')),
    ('scaler', MinMaxScaler())
])
```

```python
# создаём общий пайплайн для подготовки данных
data_preprocessor = ColumnTransformer(
    [
        ('num', MinMaxScaler(), num_columns)
    ], 
    remainder='passthrough'
)
```

```python
param_grid = [
    
    # словарь для модели RandomForestRegressor()
    {
        'models': [RandomForestRegressor(random_state=RANDOM_STATE)],
        'models__n_estimators': [50, 100, 300, 500, 700, 1000, 1500, 2000],
        'models__max_depth': [5, 7, 10, 15, 20, 30, 40],
        'models__max_features': ['auto', 'sqrt', 'log2'],
        'models__min_samples_split': [2, 3, 4, 5],
        'models__min_samples_leaf': [1, 2, 4, 6, 8, 10],
        'models__oob_score': [True, False],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },
    
    # словарь для модели CatBoostRegressor()
    {
        'models': [CatBoostRegressor(iterations=1000,
            learning_rate=0.01,
            random_state=RANDOM_STATE,
            verbose=0,
            loss_function='RMSE')],
        'models__depth': [4, 6, 8, 10],
        'models__l2_leaf_reg': [3, 5, 7, 10],
        'models__bootstrap_type': ['Bernoulli', 'MVS'],
        'models__subsample': [0.8, 0.9, 1.0],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    }
]
```

```python
# создаём итоговый пайплайн: подготовка данных и модель
pipe_final = Pipeline(
    [
        ('preprocessor', data_preprocessor),
        ('models', RandomForestRegressor())
    ]
)
```

Подберем лучшую модель.

```python
# Для подбора параметров используем RandomizedSearchCV()
random_search = RandomizedSearchCV(
    pipe_final, 
    param_grid, 
    scoring ='neg_mean_absolute_error',
    cv=3,
    random_state=RANDOM_STATE,
    n_jobs=-1,
    n_iter=5
)
random_search.fit(X_train, y_train)
```

---
Выведем параметры лучшей модели и ее метрику на кросс-валидации.

```python
print(f'Лучшая модель и её параметры:\n{random_search.best_estimator_}\n')
print(f'Метрика MAE для лучшей модели:\n{- random_search.best_score_:.3f}\n')
```

Метрика у лучшей модели на кросс-валидации превосходит пороговое значение, далее проверим ее на тестовой выборке.
